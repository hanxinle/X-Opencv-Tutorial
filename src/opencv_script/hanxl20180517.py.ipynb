```python



import cv2
import numpy as np

# Load our image as greyscale 
image = cv2.imread('images/yaohe/1.jpg',0)
b = cv2.imread('images/lj/b.jpg',0)


# Let's make our image 3/4 of it's original size
im = cv2.resize(image, None, fx=0.2, fy=0.2)
bm = cv2.resize(b,None,fx=0.2,fy=0.2)

cv2.imshow('Original', im)
cv2.waitKey() 

cv2.imshow('bm', bm)
cv2.waitKey() 

A = im - bm

cv2.imshow('Am',A)
cv2.waitKey() 

kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(A, kernel, iterations = 1)
cv2.imshow('Erosion', erosion)
cv2.waitKey()


opening = cv2.morphologyEx(A, cv2.MORPH_OPEN, kernel)
cv2.imshow('Opening', opening)
cv2.waitKey()

# Closing - Good for removing noise
closing = cv2.morphologyEx(A, cv2.MORPH_CLOSE, kernel)
cv2.imshow('Closing', closing)
cv2.waitKey()

ret,thresh1 = cv2.threshold(opening, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('1 Threshold Binary', thresh1)
cv2.waitKey()


# Values below 127 go to 255 and values above 127 go to 0 (reverse of above)
ret,thresh2 = cv2.threshold(opening, 127, 255, cv2.THRESH_BINARY_INV)
cv2.imshow('2 Threshold Binary Inverse', thresh2)
cv2.waitKey()

opening = cv2.morphologyEx(thresh2, cv2.MORPH_OPEN, kernel)
cv2.imshow('Opening-th2', opening)
cv2.waitKey()

# Values above 127 are truncated (held) at 127 (the 255 argument is unused)
ret,thresh3 = cv2.threshold(opening, 127, 255, cv2.THRESH_TRUNC)
cv2.imshow('3 THRESH TRUNC', thresh3)
cv2.waitKey()

# Values below 127 go to 0, above 127 are unchanged  
ret,thresh4 = cv2.threshold(opening, 127, 255, cv2.THRESH_TOZERO)
cv2.imshow('4 THRESH TOZERO', thresh4)
cv2.waitKey()
# Resever of above, below 127 is unchan
ged, above 127 goes to 0
ret,thresh5 = cv2.threshold(opening, 127, 255, cv2.THRESH_TOZERO_INV)
cv2.imshow('5 THRESH TOZERO INV', thresh5)
cv2.waitKey(0) 
    
    
cv2.destroyAllWindows()
```
